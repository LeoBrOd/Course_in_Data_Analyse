```python
import numpy as np

# Example survey data matrix (5 respondents, 4 questions)
survey_data = np.array([[5, 1, 4, 4],
                        [4, 2, 4, 3],
                        [5, 1, 1, 5],
                        [3, 2, 1, 2],
                        [2, 1, 2, 5]])

# Calculating various statistics
average_responses = np.mean(survey_data, axis=0)
total_responses = np.sum(survey_data, axis=0)
max_responses = np.max(survey_data, axis=0)
min_responses = np.min(survey_data, axis=0)

print("Average Responses:", average_responses)
print("Total Responses:", total_responses)
print("Maximum Responses:", max_responses)
print("Minimum Responses:", min_responses)
```

```python
import numpy as np

# Movie ratings matrix (5 viewers, 3 movies)
movie_ratings = np.array([
    [5, 1, 4],  # Viewer 1
    [4, 4, 2],  # Viewer 2
    [4, 3, 5],  # Viewer 3
    [1, 1, 5],  # Viewer 4
    [3, 2, 1]   # Viewer 5
])

average_ratings = np.mean(movie_ratings, axis=0)
print("Average Ratings:", average_ratings)

viewer_max = np.argmax(movie_ratings, axis=1)
viewer_max = viewer_max + 1
print("Viewer with Maximum Ratings:", viewer_max)
viwer_min = np.argmin(movie_ratings, axis=1)
viwer_min = viwer_min + 1
print("Viewer with Minimum Ratings:", viwer_min)
```

```python
import numpy as np
from scipy import stats

# Mock data for customer satisfaction ratings
product_a_ratings = np.array([8, 9, 7, 8, 6, 9, 7, 8, 7, 8, 9, 6, 8, 7, 9, 8, 9, 6, 7, 8, 7, 9, 8, 7, 8, 9, 6, 8, 7, 9])
product_b_ratings = np.array([6, 5, 7, 5, 6, 5, 7, 5, 6, 5, 4, 7, 5, 6, 5, 7, 5, 6, 5, 4, 7, 5, 6, 5, 7, 5, 6, 5, 4, 7])

# Task 1: Calculate Descriptive Statistics

# Product A
mean_a = np.mean(product_a_ratings)
variance_a = np.var(product_a_ratings)
skewness_a = stats.skew(product_a_ratings)
kurtosis_a = stats.kurtosis(product_a_ratings)

# Product B
mean_b = np.mean(product_b_ratings)
variance_b = np.var(product_b_ratings)
skewness_b = stats.skew(product_b_ratings)
kurtosis_b = stats.kurtosis(product_b_ratings)

print("Product A - Mean:", mean_a)
print("Product A - Variance:", variance_a)
print("Product A - Skewness:", skewness_a)
print("Product A - Kurtosis:", kurtosis_a)
print()
print("Product B - Mean:", mean_b)
print("Product B - Variance:", variance_b)
print("Product B - Skewness:", skewness_b)
print("Product B - Kurtosis:", kurtosis_b)

# Task 2: Perform a Two-Sample t-test
t_statistic, p_value = stats.ttest_ind(product_a_ratings, product_b_ratings)

print()
print("Two-Sample t-test - t-statistic:", t_statistic)
print("Two-Sample t-test - p-value:", p_value)

# Task 3: Interpretation of Results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in the average satisfaction ratings between Product A and Product B.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the average satisfaction ratings between Product A and Product B.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Data Creation
# Generate date range
date_range = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')

# Simulate daily temperature and humidity data
np.random.seed(30)
temperature_data = np.random.uniform(10, 35, size=len(date_range))
humidity_data = np.random.uniform(30, 90, size=len(date_range))

# Create DataFrame
data = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature_data,
    'Humidity': humidity_data
})
print(data.head())
```

```python
# Step 2: Data Analysis
# Calculate the mean, median, and standard deviation for both temperature and humidity
mean_temperature = data['Temperature'].mean()
median_temperature = data['Temperature'].median()
std_temperature = data['Temperature'].std()

mean_humidity = data['Humidity'].mean()
median_humidity = data['Humidity'].median()
std_humidity = data['Humidity'].std()

print(f"Temperature - Mean: {mean_temperature}, Median: {median_temperature}, Std: {std_temperature}")
print(f"Humidity - Mean: {mean_humidity}, Median: {median_humidity}, Std: {std_humidity}")
```

```python
# Identify the date with the highest temperature and the date with the highest humidity
highest_temp_date = data.loc[data['Temperature'].idxmax(), 'Date']
highest_humidity_date = data.loc[data['Humidity'].idxmax(), 'Date']
print(data['Temperature'].max())
print(data['Humidity'].max())
print(f"Date with highest temperature: {highest_temp_date}")
print(f"Date with highest humidity: {highest_humidity_date}")
```

```python

# [Bonus] Plot the temperature and humidity trends over the year
plt.figure(figsize=(12, 6))

# Plot temperature
plt.subplot(2, 1, 1)
plt.plot(data['Date'], data['Temperature'], label='Temperature')
plt.title('Daily Temperature in 2022')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()

# Plot humidity
plt.subplot(2, 1, 2)
plt.plot(data['Date'], data['Humidity'], label='Humidity', color='orange')
plt.title('Daily Humidity in 2022')
plt.xlabel('Date')
plt.ylabel('Humidity (%)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python

# Step 3: Window Function Exercise
# Use a rolling window of 7 days to calculate the moving average for both temperature and humidity
data['Temperature_MA7'] = data['Temperature'].rolling(window=10).mean()
data['Humidity_MA7'] = data['Humidity'].rolling(window=10).mean()

print(data[['Date', 'Temperature', 'Temperature_MA7', 'Humidity', 'Humidity_MA7']].head(15))
```

```python



# [Bonus] Plot the original data and the 7-day moving average on the same graph for comparison
plt.figure(figsize=(12, 6))

# Plot temperature and moving average
plt.subplot(2, 1, 1)
plt.plot(data['Date'], data['Temperature'], label='Temperature', alpha=0.5)
plt.plot(data['Date'], data['Temperature_MA7'], label='7-Day Moving Average', color='red')
plt.title('Daily Temperature and 7-Day Moving Average in 2022')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()

# Plot humidity and moving average
plt.subplot(2, 1, 2)
plt.plot(data['Date'], data['Humidity'], label='Humidity', alpha=0.5, color='orange')
plt.plot(data['Date'], data['Humidity_MA7'], label='7-Day Moving Average', color='green')
plt.title('Daily Humidity and 7-Day Moving Average in 2022')
plt.xlabel('Date')
plt.ylabel('Humidity (%)')
plt.legend()

plt.tight_layout()
plt.show()
```
