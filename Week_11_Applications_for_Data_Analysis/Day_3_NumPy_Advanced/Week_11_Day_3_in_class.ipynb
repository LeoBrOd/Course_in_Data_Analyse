3. Challenge Exercise: Analyzing Customer Satisfaction Data

  Scenario:

  Your company has conducted a customer satisfaction survey for two different products. The survey asked customers to rate their satisfaction on a scale from 1 to 10. You have the survey data for 30 customers for each product. Your task is to analyze this data.

```python
product_a_ratings = np.array([8, 9, 7, 8, 6, 9, 7, 8, 7, 8, 9, 6, 8, 7, 9, 8, 9, 6, 7, 8, 7, 9, 8, 7, 8, 9, 6, 8, 7, 9])
product_b_ratings = np.array([6, 5, 7, 5, 6, 5, 7, 5, 6, 5, 4, 7, 5, 6, 5, 7, 5, 6, 5, 4, 7, 5, 6, 5, 7, 5, 6, 5, 4, 7])
```

Tasks:
1. **Calculate Descriptive Statistics**:
   - Mean
   - Variance
   - Skewness
   - Kurtosis
2. **Perform a Two-Sample t-test**:
   - Compare the average satisfaction ratings between Product A and Product B.
3. **Interpret the Results**:
   - Discuss the skewness and kurtosis values.
   - Based on the t-test, determine if the difference in average satisfaction ratings between the two products is statistically significant.

```python
a_mean = np.mean(product_a_ratings)
b_mean = np.mean(product_b_ratings)
a_var = np.var(product_a_ratings)
b_var = np.var(product_b_ratings)
a_skew = stats.skew(product_a_ratings)
b_skew = stats.skew(product_b_ratings)
a_kurt = stats.kurtosis(product_a_ratings)
b_kurt = stats.kurtosis(product_b_ratings)
print("Product A Mean:", a_mean)
print("Product A Variance:", a_var)
print("Product A Skewness:", a_skew)
print("Product A Kurtosis:", a_kurt)
print("Product B Mean:", b_mean)
print("Product B Variance:", b_var)
print("Product B Skewness:", b_skew)
print("Product B Kurtosis:", b_kurt)
```

```python
t_statistic, p_value = stats.ttest_ind(product_a_ratings, product_b_ratings)
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
alpha=0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
