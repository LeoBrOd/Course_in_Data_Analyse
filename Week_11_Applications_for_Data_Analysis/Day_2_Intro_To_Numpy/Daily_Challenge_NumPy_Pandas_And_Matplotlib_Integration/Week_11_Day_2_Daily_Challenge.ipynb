```python
city=np.array(f"City {i}" for i in range(1,11))
month= ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temperatures=np.random.uniform(low=-5, high=35, size=(10, 12))
df=pd.DataFrame(temperatures, columns=month, index=city)
print(df)
```

```python
avg_temp=df.mean(axis=1)
low_avg_city=avg_temp.idxmin()
high_avg_city=avg_temp.idxmax()
print("City with the lowest average temperature:", low_avg_city)
print("City with the highest average temperature:", high_avg_city)
```

```python
plt.figure(figsize=(14, 7))
for city, city_temps in df.iterrows():
    plt.plot(city_temps.index, city_temps.values)
plt.xlabel("Month")
plt.ylabel("Temperature (Â°C)")
plt.title("Monthly Temperature Trends for each City")
plt.legend(df.index)
plt.grid(True)
plt.show()
```
