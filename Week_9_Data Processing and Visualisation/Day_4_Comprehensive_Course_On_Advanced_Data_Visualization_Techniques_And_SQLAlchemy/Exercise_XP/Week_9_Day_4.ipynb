# **Exercise 1 : Basic Scatter Plot With Plotnine**

```python
import pandas as pd
import numpy as np
from plotnine import *
import plotly.express as px
```

```python
data=pd.DataFrame({"x":np.random.rand(20), "y":np.random.rand(20)})
scatterPlot=(
    ggplot(data, aes(x="x", y="y"))
    + geom_line(color="blue")
    + geom_point(color="red")
    +labs(x="X-Value", y="Y-Value", title="Basic Scatter Plot")
)
print(scatterPlot)
```

# **Exercise 2 : Bar Plot Of Categorical Data With Plotnine**

```python
products=["A","B","C"]
data=pd.DataFrame({"products":np.random.choice(products,30)})
barPlot=(
    ggplot(data, aes(x="products"))
    + geom_bar(stat="count", fill="green")
    +labs(x="Products", y="Count", title="Category Frequency")
)
print(barPlot)
```

# **Exercise 3 : Line Plot With Multiple Lines With Plotnine**

```python
x=np.random.rand(20)
y1=np.sin(x)
y2=np.cos(x)
data=pd.DataFrame({"x":x, "y1":y1, "y2":y2})
linePlot=(
    ggplot(data, aes(x="x"))
    + geom_line(aes(y="y1"),color="green")
    + geom_line(aes(y="y2"),color="red")
    +labs(x="X-Value", y="Y-Value", title="Comparison of Two Functions")
    )
print(data.sort_values(by="x"))
print(linePlot)
```

# Exercise 4 : Basic Line Chart With **Plotly**

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]
data = pd.DataFrame({'months': months, 'sales': sales})
fig = px.line(data, x='months', y='sales', title='Monthly Sales Data')
fig.show()
```

# Exercise 5 : Bar Chart With Customization With **Plotly**

```python
genres = ['Fiction', 'Non-Fiction', 'Sci-Fi', 'Mystery', 'Biography']
books_sold = [120, 150, 90, 80, 130]
data = pd.DataFrame({'genres': genres, 'books_sold': books_sold})
fig = px.bar(data, x='genres', y='books_sold', title='Relationship between Age, Salary, and Experience', color="books_sold")
fig.show()
```

# **Exercise 6 : Scatter Plot With Size And Color Dimensions With Plotly**

```python
data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Years of Experience': [2, 5, 7, 10, 12]
})
fig = px.scatter(data, x='Age', y='Salary', size='Years of Experience', color='Years of Experience')
fig.show()
```
