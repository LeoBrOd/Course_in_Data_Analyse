# Exercise 2 : Scraping Robots.Txt From Wikipedia

```python
url = "https://en.wikipedia.org/robots.txt"
response=urlopen(url)
html=response.read()
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

# Exercise 4 : Checking For Page Title

```python
def check_title(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.text, 'html.parser')
  return soup.title.text
```

```python
check_title('https://en.wikipedia.org/')
```

```python
import random
```

```python
soup.find_all('li',class_="pmc-fallback-list-item-wrap lrv-u-margin-b-2")
```

```python
def generate_movie_lib(soup):
  movies = []
  movie_elements = soup.find_all('li',class_="pmc-fallback-list-item-wrap lrv-u-margin-b-2")
  for element in movie_elements:
    title = element.find('h2').text
    description = element.find('p', class_="paragraph larva // lrv-u-margin-lr-auto lrv-a-font-body-m").text.strip()
    movies.append({
            'title': title,
            'description': description
        })
  random_movies = random.sample(movies, 10)
  return random_movies

generate_movie_lib(soup)
```
