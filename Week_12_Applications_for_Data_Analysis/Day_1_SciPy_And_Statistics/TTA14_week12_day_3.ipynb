```python
from scipy import stats  # Importing the statistics module from the SciPy library
import numpy as np  # Importing the NumPy library for numerical operations
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

mu, sigma = 0, 0.9  # Setting the mean (μ) to 0 and standard deviation (σ) to 0.1
gaussian_dist = stats.norm(mu, sigma)  # Creating a Gaussian distribution object with the given μ and σ
print("PDF at x=0.5:", gaussian_dist.pdf(0.5))  # Calculating and printing the probability density at x=0.5

x_gaussian = np.linspace(-0.5, 0.5, 100)  # Generating 100 evenly spaced values from -0.5 to 0.5
pdf_gaussian = gaussian_dist.pdf(x_gaussian)  # Calculating the PDF for each value in x_gaussian
plt.figure(figsize=(6, 4))  # Creating a figure for the plot with specified size
plt.plot(x_gaussian, pdf_gaussian)  # Plotting the PDF against x values
plt.title('Gaussian Distribution\n$\mu=0$, $\sigma=0.1$')  # Adding a title with mean and standard deviation
plt.xlabel('x')  # Labeling the x-axis
plt.ylabel('PDF')  # Labeling the y-axis
plt.show()  # Displaying the plot
```

```python
# Plotting Binomial Distribution
x_binom = np.arange(0, n+1)  # Generating values from 0 to n (inclusive)
pmf_binom = binom_dist.pmf(x_binom)  # Calculating the PMF for each value in x_binom
plt.figure(figsize=(6, 4))  # Creating a figure for the plot with specified size
plt.bar(x_binom, pmf_binom)  # Creating a bar plot for the PMF
plt.title('Binomial Distribution\n$n=10$, $p=0.5$')  # Adding a title with number of trials and probability of success
plt.xlabel('Number of Successes')  # Labeling the x-axis
plt.ylabel('PMF')  # Labeling the y-axis
plt.show()  # Displaying the plot
```

```python
import numpy as np  # Importing the NumPy library for numerical operations
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting
from scipy import stats  # Importing the statistics module from the SciPy library

# Generate sample data
np.random.seed(0)  # Setting a random seed for reproducibility
group1 = stats.norm.rvs(loc=5, scale=10, size=100)  # 100 data points, mean=5, std=10
group2 = stats.norm.rvs(loc=5, scale=10, size=100)  # 100 data points, mean=5, std=10

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(group1, group2)
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Visualization
plt.figure(figsize=(10, 6))
plt.boxplot([group1, group2], labels=['Group 1', 'Group 2'])
plt.title('Comparison Between Two Groups')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Combine all groups for PDF visualization
all_weights = data['Weight']
mean_weight = np.mean(all_weights)
std_weight = np.std(all_weights)

# Generate a range of values for plotting the PDF
x = np.linspace(min(all_weights), max(all_weights), 100)
pdf = norm.pdf(x, mean_weight, std_weight)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='PDF')
plt.hist(all_weights, bins=20, density=True, alpha=0.6, color='g', label='Histogram of Weight')
plt.xlabel('Weight')
plt.ylabel('Density')
plt.title('Probability Density Function (PDF) of Plant Weight')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a histogram to approximate the PMF
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(all_weights, bins=20, density=True, alpha=0.6, color='b', label='Histogram (PMF Approximation)')
bin_centers = 0.5 * (bins[1:] + bins[:-1])
plt.stem(bin_centers, counts, 'b', markerfmt='bo', basefmt='-', use_line_collection=True)
plt.xlabel('Weight')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) Approximation of Plant Weight')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np  # Importing the NumPy library for numerical operations
from scipy import stats  # Importing the statistics module from the SciPy library
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Example data
x = np.array([5, 10, 15, 20, 25])
y = np.array([12, 24, 37, 43, 50])

# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# The slope and intercept of the regression line
print("Slope:", slope)
print("Intercept:", intercept)
print("R-value (Correlation Coefficient):", r_value)
print("P-value:", p_value)

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Data points')  # Plotting the data points
plt.plot(x, slope * x + intercept, color='red', label='Regression line')  # Plotting the regression line
plt.title('Linear Regression Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np  # Importing the NumPy library for numerical operations
from scipy import stats  # Importing the statistics module from the SciPy library
import matplotlib.pyplot as plt  # Importing Matplotlib for plotting

# Sample data from three different groups
group1 = [20, 21, 19, 22, 24]
group2 = [28, 30, 27, 26, 29]
group3 = [33, 35, 34, 31, 32]

# Performing one-way ANOVA
f_value, p_value = stats.f_oneway(group1, group2, group3)
print("F-value:", f_value)
print("P-value:", p_value)

# Visualization
plt.figure(figsize=(10, 6))
plt.boxplot([group1, group2, group3], labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('ANOVA: Comparison of Three Groups')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Generate sample data
np.random.seed(0)
x = np.random.normal(50, 10, 100)
y = 2 * x + np.random.normal(0, 10, 100)

# Calculate Pearson correlation coefficient
pearson_corr, pearson_p_value = stats.pearsonr(x, y)
print("Pearson Correlation Coefficient:", pearson_corr)
print("Pearson P-value:", pearson_p_value)

# Calculate Spearman correlation coefficient
spearman_corr, spearman_p_value = stats.spearmanr(x, y)
print("Spearman Correlation Coefficient:", spearman_corr)
print("Spearman P-value:", spearman_p_value)

# Calculate covariance
cov_matrix = np.cov(x, y)
cov_xy = cov_matrix[0, 1]
print("Covariance:", cov_xy)

# Visualization - Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, label='Data points')
sns.regplot(x=x, y=y, scatter=False, color='red', label='Regression line')
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.grid(True)
plt.show()

# Visualization - Correlation matrix heatmap
data = pd.DataFrame({'X': x, 'Y': y})
corr_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Generate example data
np.random.seed(0)
advertising_spend = np.random.normal(1000, 200, 100)  # Mean: 1000, Std: 200
revenue = 2.5 * advertising_spend + np.random.normal(0, 200, 100)  # Linear relationship with noise

# Calculate Pearson correlation
pearson_corr, pearson_p_value = pearsonr(advertising_spend, revenue)
print(f'Pearson Correlation Coefficient: {pearson_corr:.2f}')
print(f'Pearson P-value: {pearson_p_value:.2e}')

# Plot the data with a regression line
plt.figure(figsize=(10, 6))
plt.scatter(advertising_spend, revenue, label='Data points')
plt.plot(advertising_spend, np.poly1d(np.polyfit(advertising_spend, revenue, 1))(advertising_spend), color='red', label='Regression line')
plt.title(f'Pearson Correlation (r={pearson_corr:.2f})')
plt.xlabel('Advertising Spend ($)')
plt.ylabel('Revenue ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Generate example ordinal data
np.random.seed(0)
customer_satisfaction = np.random.randint(1, 6, 50)  # Ratings from 1 (Very Unsatisfied) to 5 (Very Satisfied)
service_quality = np.random.randint(1, 6, 50)  # Ratings from 1 (Poor) to 5 (Excellent)

# Calculate Spearman correlation
spearman_corr, spearman_p_value = spearmanr(customer_satisfaction, service_quality)
print(f'Spearman Correlation Coefficient: {spearman_corr:.2f}')
print(f'Spearman P-value: {spearman_p_value:.2e}')

# Rank the data for visualization
customer_satisfaction_ranked = pd.Series(customer_satisfaction).rank()
service_quality_ranked = pd.Series(service_quality).rank()

# Plot the data with ranks
plt.figure(figsize=(10, 6))
plt.scatter(customer_satisfaction_ranked, service_quality_ranked, label='Ranked data points')
plt.plot(customer_satisfaction_ranked, np.poly1d(np.polyfit(customer_satisfaction_ranked, service_quality_ranked, 1))(customer_satisfaction_ranked), color='red', label='Regression line')
plt.title(f'Spearman Correlation (r={spearman_corr:.2f})')
plt.xlabel('Ranked Customer Satisfaction')
plt.ylabel('Ranked Service Quality')
plt.legend()
plt.grid(True)
plt.show()
```
