Exercise 3 : Visualizing Monthly Sales With A Bar Chart

```python
monthly_sales = [5000, 5500, 6200, 7000, 7500,5000, 5500, 6200, 7000, 7500,5000, 7500]
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.bar(months, monthly_sales)
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
sales_data= pd.read_csv('sales_data.csv')
```

```python
sales_data.head()
```

Calculate the total quantity of products sold by the company.


```python
total_quantity = sales_data['quantity'].sum()
print("Total quantity of products sold by the company:", total_quantity)
```


Identify the category that had the highest revenue and calculate how much revenue it generated.

```python
category_revenue = sales_data.groupby('category')['revenue'].sum()
highest_revenue_category = category_revenue.idxmax()
highest_revenue_amount = category_revenue.max()
print("Category with highest revenue:", highest_revenue_category)
print("Revenue generated by", highest_revenue_category, ":", highest_revenue_amount)
```


Determine the average revenue per sale in the dataset.


```python
average_revenue = sales_data['revenue'].mean()
print("Average revenue per sale:", average_revenue)
```


Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

```python
sales_data['date'] = pd.to_datetime(sales_data['date'])
sales_data['quarter'] = sales_data['date'].dt.quarter
quarter_revenue = sales_data.groupby('quarter')['revenue'].sum()
quarter_revenue.plot(kind='bar', color=['skyblue', 'lightgreen', 'gold', 'coral'])
plt.xlabel('Quarter')
plt.ylabel('Total Revenue')
plt.title('Revenue Generated by Quarter')
plt.show()
```

Exercise 5 : Data Visualisation Using MatPlotLib

Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.
Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.
Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.`

```python
import numpy as np
```

```python
x = np.linspace(-10, 10, 100)
y = x**2
plt.plot(x, y)
plt.title('Line Plot: y = x^2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]
plt.bar(products, sales)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Weekly Sales of Products')
plt.tight_layout()
plt.show()
```

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
percentages = [40, 30, 20, 10]
plt.pie(percentages, labels=fruits, autopct="%1.1f%%", startangle=140)
colors = ['lightcoral', 'gold', 'lightblue', 'lightgreen']
plt.gca().set_prop_cycle('color', colors)
plt.legend(fruits, loc='best')
plt.title('Favorite Fruits Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
