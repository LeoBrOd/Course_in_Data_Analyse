Preparing data

```python
women= (mhealth_data['Gender'] == 'female') | (mhealth_data['Gender'] == 'F') | (mhealth_data['Gender'] == 'Femail') | (mhealth_data['Gender'] == 'f') | (mhealth_data['Gender'] == 'Femake') | (mhealth_data['Gender'] == 'Woman') | (mhealth_data['Gender'] == 'Female ')  | (mhealth_data['Gender'] == 'femail') | (mhealth_data['Gender'] == 'Cis Female') | (mhealth_data['Gender'] == 'cis-female/femme') | (mhealth_data['Gender'] == 'Female (cis)')
men = (mhealth_data['Gender'] == 'M') | (mhealth_data['Gender'] == 'male') | (mhealth_data['Gender'] == 'm') | (mhealth_data['Gender'] == 'maile') | (mhealth_data['Gender'] == 'Mal') | (mhealth_data['Gender'] == 'Make') | (mhealth_data['Gender'] == 'Male ') | (mhealth_data['Gender'] == 'Man') | (mhealth_data['Gender'] == 'msle') | (mhealth_data['Gender'] == 'msle') | (mhealth_data['Gender'] == 'Mail') | (mhealth_data['Gender'] == 'Malr') | (mhealth_data['Gender'] == 'Cis Male') | (mhealth_data['Gender'] == 'Male (CIS)') | (mhealth_data['Gender'] == 'cis male') | (mhealth_data['Gender'] == 'Cis Man')
mhealth_data.loc[women, 'Gender'] = 'Female'
mhealth_data.loc[men, 'Gender'] = 'Male'
mhealth_data.loc[(mhealth_data['Gender'] != 'Female') & (mhealth_data['Gender'] != 'Male'), 'Gender'] = 'Non-binary'
```

```python
mhealth_data.drop(mhealth_data[(mhealth_data['Age']<18) | (mhealth_data['Age']>66)].index, inplace=True)
```

```python
data=mhealth_data[['Age', 'Gender', 'Country', 'treatment']]
```

```python
data.head()
```

Answers

1. What is the distribution of mental health conditions among different age groups in the tech industry?

```python
data['age_group'] = pd.cut(data['Age'], bins=[18, 25, 35, 45, 55, 67], labels=['18-24', '25-34', '35-44', '45-54', '55-66'])
mental_health_by_age = data.groupby('age_group')['treatment'].value_counts().unstack(fill_value=0)
```

```python
mental_health_by_age.plot(kind='bar', stacked=False)
plt.xlabel('Age Group')
plt.ylabel('Number of Individuals')
plt.title('Distribution of Mental Health Treatment by Age Group')
plt.show()
```

2. How does the frequency of mental health issues vary by gender?

```python
mental_health_by_gender = data.groupby('Gender')['treatment'].value_counts()
total_per_gender = data['Gender'].value_counts()
mental_health_percentages = mental_health_by_gender / total_per_gender * 100
mental_health_by_gender.plot(kind='bar', stacked=False)
plt.xlabel('Gender')
plt.ylabel('Number of Individuals')
plt.title('Frequency of Mental Health Issues by Gender')
plt.show()
```

3. Identify the countries with the highest and lowest reported rates of mental health issues in the tech industry.

```python
treatment = data[data['treatment']=='Yes']
mental_health_by_country = treatment.loc[:, ['treatment', 'Country']].groupby('Country').count()
total_per_country = data['Country'].value_counts()
sorted_countries = total_per_country.sort_values(ascending=False)
highest_rate_country = sorted_countries.index[0]
lowest_rate_country = sorted_countries.index[-1]
print("Country with the highest rate of mental health issues:", highest_rate_country)
print("Country with the lowest rate of mental health issues:", lowest_rate_country)
mental_health_by_country.plot(kind='bar', stacked=False)
plt.xlabel('Country')
plt.ylabel('Number of Individuals')
plt.title('Reported rates of Mental Health Issues by Country')
plt.show()
```
