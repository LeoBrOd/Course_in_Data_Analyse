Is there a significant difference in quality between red and white wines?

```python
white_wine_total=white_red_wine[white_red_wine['type']=='white']['quality'].count()
white_wine=white_red_wine[white_red_wine['type']=='white']['quality'].value_counts()/white_wine_total*100
white_wine_mean=white_red_wine[white_red_wine['type']=='white']['quality'].mean()
print("White wine mean",white_wine_mean)

red_wine_total=white_red_wine[white_red_wine['type']=='red']['quality'].count()
red_wine=white_red_wine[white_red_wine['type']=='red']['quality'].value_counts()/red_wine_total*100
red_wine_mean=white_red_wine[white_red_wine['type']=='red']['quality'].mean()
print("Red wine mean",red_wine_mean)
print("White wine",white_wine)
print("Red wine",red_wine)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x=white_wine.index, y=white_wine.values, color='skyblue', label='White Wine')
sns.barplot(x=red_wine.index, y=red_wine.values, color='lightcoral', label='Red Wine')
plt.xlabel('Quality Rating')
plt.ylabel('Percentage (%)')
plt.title('Wine Quality Ratings')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

Differnce in my opinion is not significant, but white wines have few higher marks, such as "9" and little higher percent of "6,7,8" marks
