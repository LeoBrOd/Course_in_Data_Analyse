```python
import pandas as pd

data = {
    'Book Title': ['The Great Gatsby', 'To Kill a Mockingbird', '1984', 'Pride and Prejudice', 'The Catcher in the Rye'],
    'Author': ['F. Scott Fitzgerald', 'Harper Lee', 'George Orwell', 'Jane Austen', 'J.D. Salinger'],
    'Genre': ['Classic', 'Classic', 'Dystopian', 'Classic', 'Classic'],
    'Price': [10.99, 8.99, 7.99, 11.99, 9.99],
    'Copies Sold': [500, 600, 800, 300, 450]
}

df = pd.DataFrame(data)
```

```python
print(df.head())
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
df['Total Revenue']=df['Price']*df['Copies Sold']
print(df)
```

```python
highest_book=df.loc[df['Total Revenue'].idxmax()]
highest_book
```

```python
lowest_book=df.loc[df['Total Revenue'].idxmin()]
lowest_book
```

```python
genre_stats=df.groupby('Genre').agg({'Price':'mean','Copies Sold':'sum'}).reset_index()
genre_stats
```

```python
filtered_books=df[(df['Genre']=='Classic')&(df['Copies Sold']>400)]
filtered_books
```
