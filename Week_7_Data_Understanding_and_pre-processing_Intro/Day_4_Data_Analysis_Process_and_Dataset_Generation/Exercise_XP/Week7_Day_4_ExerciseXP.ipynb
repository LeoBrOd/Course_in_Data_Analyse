Exercise 4 : Synthetic Data Generation

```python
import pandas as pd
import numpy as np
from faker import Faker
```

```python
fake = Faker()

names = [fake.name() for _ in range(100)]
addresses = [fake.address() for _ in range(100)]
emails = [fake.email() for _ in range(100)]
ages = np.random.randint(20, 61, size=100)
incomes = np.random.randint(7000, 15000, size=100)

data = {'Name': names, 'Address': addresses, 'Email': emails, 'Age': ages, 'Income': incomes}
df = pd.DataFrame(data)

print(df.head(10))
```
